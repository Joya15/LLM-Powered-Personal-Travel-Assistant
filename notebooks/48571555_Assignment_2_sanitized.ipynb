```python
import os
import cohere
from dotenv import load_dotenv

# Load the API key from a local .env file or environment variable.
load_dotenv()
api_key = os.getenv("COHERE_API_KEY")
if not api_key:
    raise ValueError("Set COHERE_API_KEY before running this notebook.")

co = cohere.Client(api_key)
```

```python
response1 = co.generate(
    model='command-r-plus',
    prompt="Plan a 5-day budget-friendly trip to Sydney for a student interested in nature and street food.",
    max_tokens=800,
    temperature=0.7
)

print("Response 1:\n", response1.generations[0].text)
```

```python
prompts = [
    "Suggest a 3-day beach trip from Sydney under $1000 AUD.",
    "Plan a romantic weekend getaway in Italy.",
    "Give me a solo travel plan across Southeast Asia for 2 weeks.",
    "List popular vegetarian-friendly cities to visit in Europe."
]

for i, p in enumerate(prompts, 1):
    response = co.generate(
        model='command-r-plus',
        prompt=p,
        max_tokens=800,
        temperature=0.7
    )
    print(f"\nPrompt {i}: {p}")
    print(f"Response {i}:\n{response.generations[0].text.strip()}")
```

## 2. Assistant Dialogue System

```python
def get_response(prompt):
    response = co.generate(
        model='command-r-plus',
        prompt=prompt,
        max_tokens=1800,
        temperature=0.7
    )
    return response.generations[0].text.strip()
```

```python
name = input("What's your name? ")
destination = input("Where do you want to travel? ")
days = input("How many days will you be staying? ")
interest = input("What kind of travel do you prefer (e.g., nature, food, culture)? ")

user_profile = f"My name is {name}. I want to go to {destination} for {days} days. I'm interested in {interest}."
prompt = f"Based on this info, plan a personalized travel itinerary:\n{user_profile}"

response = get_response(prompt)
print("\n--- Personalized Travel Plan ---")
print(response)
```

```python
budget = input("What's your estimated budget? ")
climate = input("What kind of climate do you prefer (e.g., tropical, cold, temperate)? ")
companions = input("Are you traveling solo or with someone? ")

extended_profile = (
    f"{user_profile}. My estimated budget is {budget}, I prefer {climate} weather, "
    f"and I'm traveling with {companions}."
)

prompt = f"Using this info, create a detailed, budget-aware itinerary:\n{extended_profile}"
response = get_response(prompt)
print("\n--- Extended Personalized Travel Plan ---")
print(response)
```

```python
for temp in [0.3, 0.7, 0.9]:
    print(f"\n--- Prompt with temperature {temp} ---")
    res = co.generate(prompt=extended_profile, model='command-r-plus', max_tokens=300, temperature=temp)
    print(res.generations[0].text.strip())
```

# Explanation of Code

In this assignment I have created a personal travel assistant application powered by Cohere’s `command-r-plus` large language model (LLM). The assistant is designed to interact with users, gather key travel-related information and then generate personalized travel itineraries.


I have implemented this assistant in Python using the Cohere API. It simulates a conversation using `input()` prompts in a Jupyter Notebook, where the user is asked a series of questions like Name, Travel destination, Number of travel days etc.


All of the collected inputs are combined into a single prompt and then passed to the Cohere LLM using the `.generate()` function. This function returns a travel plan that reflects the user's preferences in detail.

The assistant also supports tuning creativity and specificity by adjusting the `temperature` parameter. Higher values (0.9) result in more creative outputs, while lower values ( 0.3) generate more focused and factual responses.

## Explanation of Results

I have tested several prompt scenarios to evaluate the assistant’s versatility:

- **Food-focused 10-day trip to Thailand**: The assistant generated a detailed multi-city itinerary including Bangkok, Chiang Mai, Pai, and Phuket, with relevant activities. Also, the budget, weather preference and travel style solo or group were reflected accurately.

- **3-day beach trip under $1000 AUD**: Itineraries were cost-conscious and included destinations like Port Stephens, Nelson Bay, and Stockton Sand Dunes, along with accommodation and activity suggestions within budget.
